```python
def receita():
  ingredientes = []
  modopreparo = []
  cont = 1
  cont1 = 1
  nome = input("digite o nome da receita: ").capitalize()

  while True:
    item = input(f"Digite o {cont}° ingrediente necessario para realizar {nome}, ou digite sair: ")
    cont = cont+1
    ingredientes.append(item)
    if item =="sair":
      ingredientes.remove("sair")
      break

  print("lista de ingredientes é a seguinte:")
  for i in ingredientes:
    print("*",i)


  while True:
    etapapreparo = input(f"Digite a {cont1}° etapa do preparo ou digite sair: ")
    cont1=cont1+1
    modopreparo.append(etapapreparo)
    if etapapreparo == "sair":
      modopreparo.remove("sair")
      break
  for i in modopreparo:
    print("*",i)
```

```python
receita()
```

```python
def notassemestre():
  nota1 = float(input("Digite a primeira nota: "))
  nota2 = float(input("Digite a segunda nota:"))
  nota3 = float(input("Digite a terceita nota:"))
  total = nota1+nota2+nota3
  return total

```

# com parametro e retorno

```python
def calcular_total(preco_unitario, quantidade):
  total = preco_unitario * quantidade
  return total

valor = calcular_total(25.5, 3)
print(f"O valor total da compra é de r${valor:.2f}")
```

```python
def calcular_imc(peso, altura):
  imc = peso / (altura**2)
  return imc

peso = float(input("Digite seu peso (kg): "))
altura = float(input("Digite sua altura (m): "))

resultado = calcular_imc(peso, altura)
print(f"seu imc é {resultado:.2f}")
```

```python
def calcular_soma_e_media(a,b,c):
  soma = a+b+c
  media = soma/3
  return soma,media

n1 = float(input("digite um numero: "))
n2 = float(input("digite um numero: "))
n3 = float(input("digite um numero: "))

resultado_soma, resultado_media = calcular_soma_e_media(n1,n2,n3)

print(f"A soma é {resultado_soma:.2f}")
print(f"A media é {resultado_media:.2f}")
```

```python
def calcular_desconto(valor_original,percentual_desconto):
  percentual_desconto = percentual_desconto/100
  valor_final = valor_original*(1-percentual_desconto)
  return valor_final

preco = float(input("Digite o preço do produto: "))
desconto = float(input("Digite o percentual de desconto: "))

print(f"O preço com o desconto fica R${calcular_desconto(preco, desconto)}")
```

```python
def encontrar_extremos(lista):
    if not lista:
        return None, None

    maior = max(lista)
    menor = min(lista)
    return maior, menor

lista = []

while True:
  numero = int(input("Digite um numero e 0 para sair: "))
  if numero == 0:
    break

  lista.append(numero)

print(f"o maior e menor numero da sua lista é: {encontrar_extremos(lista)}")
```

```python
def calculadora(a, b, operacao):
    operacao = operacao.lower().strip()

    if operacao == 'soma':
        return a + b
    elif operacao == 'subtracao':
        return a - b
    elif operacao == 'multiplicacao':
        return a * b
    elif operacao == 'divisao':
        if b == 0:
            return "Erro: Divisão por zero não é permitida"
        return a / b
    else:
        return "Erro: Operação inválida. Use: 'soma', 'subtracao', 'multiplicacao' ou 'divisao'"

a = int(input("Digite um numero: "))
b = int(input("Digite outro numero: "))
operacao = str(input("Digite a operação: "))

print(f"O final do calculo é {calculadora(a,b,operacao)}")
```
