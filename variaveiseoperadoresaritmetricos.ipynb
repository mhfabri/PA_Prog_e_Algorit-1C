#Conversão de Moeda

Este exemplo demonstra como o fluxograma de conversão de moeda pode ser implementado em Python utilizando a entrada do usuário, a cotação do dólar e a operação de divisão

conversaomoeda.drawio.png

```python

cotacao = float(input("Coloque o valor do dolar $"))
valorreal = float(input("coloque o valor em reais que deseja converter R$ "))


resultado = valorreal/cotacao


print(f"A conversão fica: ${resultado:.2f}")
```

Outro exemplo é o cálculo de comissão. O fluxograma é convertido em código Python, utilizando o preço unitário, a quantidade vendida e o cálculo da comissão

image.png

##Convesão de minutos em segundo
Faça um algoritmo que faça conversão de minutos e horas em segundos. Solicite ao usuário a hora e os minutos separado

conversaotempo.drawio.png

```python
horas = float(input("Quais as horas? Horas: "))
minutos = float(input("Quais os minutos? Minutos: "))
minutosdec = minutos/60
tempodec = (horas + minutosdec)*60
segundos = tempodec*60
print(f"O tempo em segundo é igual á: {segundos}")
```

##Divisão de Conta em um Restaurante com Gorjeta

Um grupo de amigos foi a um restaurante e deseja dividir a conta igualmente entre eles. Além disso, o restaurante cobra uma taxa de serviço de 10% sobre o valor total da conta mais uma gorjeta.

image.png

```python

valorconta = float(input(" Digite o valor da sua conta R$ "))
num_pessoas = float(input(" Digite por quantas pessoas iram dividir a conta: "))
valorgorjeta = float(input(" Digite o valor da gorjeta: R$ "))


taxaservico = (valorconta*10/100) + valorconta + valorgorjeta
totalporpessoa = taxaservico / num_pessoas


print(f" O valor total da conta com a gorjeta foi R$ {taxaservico:.2f} e cada pessoa vai pagar R$ {totalporpessoa:.2f}")
```
